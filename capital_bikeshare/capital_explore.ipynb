### Build models for each station

```python
models = {}
```

```python
for station in df.columns.values[:198]:
    models[station] = SVR(gamma=0.1).fit(X,df[station].values)
```

### Build predicted use at 70 degrees for clustering

```python
Xkmean = np.ndarray((198,24*7))
stationIDs = ['']*198
for indx,station in enumerate(models.keys()):
    Xkmean[indx] = models[station].predict(zip(arange(24*7),[70]*24*7,[1]*24*7))
    stationIDs[indx] = station
```

```python
imgplot = imshow(Xkmean)
imgplot.set_cmap('Greys')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeanmodel = KMeans(n_clusters=3)
```

```python
kmeanmodel.fit(Xkmean)
```

```python
groups = kmeanmodel.predict(Xkmean)
```

```python
Xkmean_ordered = np.ndarray((198,24*7))
order_indx = 0
for i in xrange(groups.max()+1):
    for line_indx in xrange(len(Xkmean)):
        if groups[line_indx] == i:
            Xkmean_ordered[order_indx] = Xkmean[line_indx]
            order_indx += 1
```

```python
imgplot = imshow(Xkmean_ordered)
imgplot.set_cmap('Greys')
```

### Export stationID and grouping for viz

```python
import csv
```

```python
with open('station_groupings.csv','w') as outfile:
    filewriter = csv.writer(outfile,delimiter=',')
    for indx in xrange(len(models)):
        filewriter.writerow([stationIDs[indx]]+[groups[indx]])
```
